```python
import scipy.stats as stats
Fs = 20
frame_size = Fs*4 # 80
hop_size = Fs*2 # 40
```

```python
def get_frames(df, frame_size, hop_size):

    N_FEATURES = 6

    frames = []
    labels = []
    for i in range(0, len(df) - frame_size, hop_size):
        acc_x = df['acc_x'].values[i: i + frame_size]
        acc_y = df['acc_y'].values[i: i + frame_size]
        acc_z = df['acc_z'].values[i: i + frame_size]
        gyro_x = df['gyro_x'].values[i: i + frame_size]
        gyro_y = df['gyro_y'].values[i: i + frame_size]
        gyro_z = df['gyro_z'].values[i: i + frame_size]
        
        
        # Retrieve the most often used label in this segment
        label = stats.mode(df['label'][i: i + frame_size])[0][0]
        frames.append([acc_x, acc_y, acc_z,gyro_x,gyro_y,gyro_z])
        labels.append(label)

    # Bring the segments into a better shape
    frames = np.asarray(frames).reshape(-1, frame_size, N_FEATURES)
    labels = np.asarray(labels)

    return frames, labels

X, y = get_frames(scaled_X, frame_size, hop_size)

X.shape, y.shape
```

```python
y=np.asarray(pd.get_dummies(y),dtype=np.float32)
```

```python
from keras.layers import LSTM
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
from kerastuner.tuners import RandomSearch
from tensorflow.keras.layers import Activation
```

```python
def build_model(hp):
    n_timesteps, n_features, n_outputs = X_train.shape[1], X_train.shape[2], y_train.shape[1]
    model2 = tf.keras.Sequential()
    
    model2.add(layers.LSTM(units=hp.Int('input_units_',min_value=100,max_value=1000,step=50), input_shape=(n_timesteps, n_features),return_sequences=True))
    model2.add(Activation('relu'))
    
    for i in range (hp.Int('num_layers', 1,40)):
            model2.add(layers.LSTM(units=hp.Int('lstm_' + str(i),min_value=100,max_value=1000,step=50),input_shape=(n_timesteps, n_features),return_sequences=True))
            model2.add(Activation('relu'))
            
            for j in np.arange (hp.Float('dropout',0,0.5,0.1)):
                model2.add(layers.Dropout(j))
    model2.add(layers.LSTM(units=hp.Int('last_lstm_unit_',min_value=100,max_value=1000,step=50),input_shape=(n_timesteps, n_features)))
    model2.add(Activation('relu')) 
    

    for i in range (hp.Int('num_layers_dense', 2,10)):
        model2.add(layers.Dense(units=hp.Int('dense_' + str(i),min_value=100,max_value=1000,step=50)))
        model2.add(Activation('relu'))

            
    model2.add(layers.Dense(n_outputs, activation='softmax'))
    model2.compile(loss='categorical_crossentropy', optimizer=tf.keras.optimizers.Adam(hp.Choice('learning_rate', [1e-2,1e-3,1e-4])), metrics=['accuracy'])
    return model2    
```

```python
tuner=RandomSearch(
                build_model,
                objective='val_accuracy',
                max_trials =5,
                executions_per_trial=3,
                directory='project',
                project_name='fall detection and prediction')
```
